```python
import cv2 as cv
import numpy as np
```

```python
import tensorflow as tf
from tensorflow import keras
best_model = keras.models.load_model("best_model.h5")
```

```python
# Capture the video
import winsound
frequency = 3000
duration = 3000
face_classifier = cv.CascadeClassifier("lbp_frontal_face_cascade.xml")
counter = 0
capture = cv.VideoCapture(0)
while True:
    isTrue, frame = capture.read()
    
# Get the eyes from the frame
    eye_classifier = cv.CascadeClassifier("haar_eye_cascade.xml")
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    
    eyesbox = eye_classifier.detectMultiScale(gray, 1.1, 4)
    for (x,y,w,h) in eyesbox:
        gray_roi = gray[y:y+h,x:x+w]
        color_roi = frame[y:y+h,x:x+w]
#         cv.rectangle(frame, (x,y), (x+w,y+w), (0,255,0),2)
        
        eyess = eye_classifier.detectMultiScale(gray_roi)
        if len(eyess) == 0:
            print("Eyes not detected!")
        else:
            for (ex,ey,ew,eh) in eyess:
                eyes_roi = color_roi[ey:ey+eh, ex:ex+ew]
                final_img = cv.resize(eyes_roi, (224,224))
                final_img = np.expand_dims(final_img, axis = 0)
                final_img = final_img/255.0
        
# Perform predictions
    pred = best_model.predict(final_img)
    if pred>0.7:
        counter += 1
        status = "Drowsy"
        gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
        print(face_classifier.empty())
        facebox = face_classifier.detectMultiScale(gray, 1.1, 6)
        for (x,y,w,h) in facebox:
            cv.rectangle(frame, (x,y), (x+w,y+h), (0,0,255), 2)
            cv.putText(frame, status, (50,50), cv.FONT_HERSHEY_SIMPLEX, 3, (0,0,255),2)
            
        if counter > 15:
            winsound.Beep(frequency, duration)
            counter = 0
    else:
        status = "Active"
        gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
        print(face_classifier.empty())
        facebox = face_classifier.detectMultiScale(gray, 1.1, 6)
        for (x,y,w,h) in facebox:
            cv.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2)
            cv.putText(frame, status, (50,50), cv.FONT_HERSHEY_SIMPLEX, 3, (0,255,0),2)
    
    cv.imshow("Drowsiness Detection", frame)
    
    if cv.waitKey(2) & 0xFF == ord("c"):
        break
        
capture.release()
cv.destroyAllWindows()
```
